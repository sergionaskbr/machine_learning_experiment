# AFFINITY PROPAGATION

## 0. Imports

```python
import numpy as np
import pandas as pd
from sklearn import cluster as ct
from sklearn import metrics as mt
from matplotlib import pyplot as plt
```

## 1. Load Dataset

```python
dataset_path = '../../datasets/clustering/X_dataset.csv'
x_dataset = pd.read_csv(dataset_path)
```

## 2. Model Training and Performance

```python
p = np.arange(-61, -1, 1)
ss_list = []

for i in p:
    affinity = ct.AffinityPropagation(preference=i)
    affinity.fit(x_dataset)
    labels = affinity.predict(x_dataset)

    ss = mt.silhouette_score(x_dataset, labels)
    ss_list.append(ss)

plt.plot(p, ss_list, '-o')
plt.xlabel('preference value')
plt.ylabel('Average Silhouette Score')
plt.show();

best_p = p[np.argmax(ss_list)]
best_p_value = ss_list[np.argmax(ss_list)]

print(f"Best P value: {best_p}")
print(f"Silhouette Score for best P: {best_p_value:.4f}")
print('Estimated number of clusters: {}'.format(len(np.unique(labels))))
```
