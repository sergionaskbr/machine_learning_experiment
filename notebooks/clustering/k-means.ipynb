# K-MEANS

## 0. Imports

```python
from matplotlib import pyplot as plt
from sklearn import cluster as ct
from sklearn import metrics as mt
import numpy as np
import pandas as pd
```

## 1. Load Dataset

```python
dataset_path = '../../datasets/clustering/X_dataset.csv'
x_dataset = pd.read_csv(dataset_path)
```

## 2. Model Training and Performance

```python
k = np.arange(2, 12, 1)
ss_list = []

for i in k:
    kmeans = ct.KMeans(n_clusters=i)
    labels = kmeans.fit_predict(x_dataset)

    ss = mt.silhouette_score(x_dataset, labels)
    ss_list.append(ss)

plt.plot(k, ss_list, '-o')
plt.xlabel('K value')
plt.ylabel('Average Silhouette Score')
plt.show();

best_k = k[np.argmax(ss_list)]
best_ss_value = ss_list[np.argmax(ss_list)]

print(f"Best K value: {best_k}")
print(f"Silhouette Score for best K: {best_ss_value:.4f}")
```
